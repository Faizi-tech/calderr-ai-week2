# What is Prompt Engineering?

Prompt Engineering is the process of designing clear, specific, and well-structured instructions (prompts) that help Large Language Models (LLMs) produce accurate, relevant, and reliable responses.

A good prompt provides enough context, specifies the task clearly, and may include constraints such as the desired format, audience, tone, or examples.

### Characteristics of a Good Prompt

- Clear objective
- Sufficient context
- Specific instructions
- Defined output format
- Appropriate constraints

# Chain of Thought (CoT)

Chain of Thought (CoT) prompting encourages the model to reason through a problem step by step before producing the final answer.

Instead of asking only for the answer, the prompt asks the model to explain its reasoning process.

### Why use CoT?

- Improves reasoning on complex tasks
- Reduces logical mistakes
- Produces more transparent answers
- Works especially well for mathematics and logic problems

### Example

Without CoT:

Question:
If Ali has 12 apples and gives away 5, how many remain?

Answer:
7

With CoT:

Question:
If Ali has 12 apples and gives away 5, think step by step before answering.

Reasoning:

- Initial apples = 12
- Apples given away = 5
- Remaining = 12 − 5 = 7

Final Answer:
7

```python
normal_prompt = """
If Ali has 12 apples and gives away 5, how many remain?
"""

cot_prompt = """
If Ali has 12 apples and gives away 5,
think step by step before answering.
"""

print("Normal Prompt:")
print(normal_prompt)

print("\n---------------------\n")

print("Chain of Thought Prompt:")
print(cot_prompt)
```

# Tree of Thoughts (ToT)

Tree of Thoughts (ToT) extends the idea of Chain of Thought by allowing the model to explore multiple reasoning paths before selecting the best one.

Instead of following a single reasoning chain, the model considers several possible solutions, evaluates them, and chooses the most appropriate answer.

Tree of Thoughts is useful for planning, puzzle solving, coding, and complex decision-making tasks.

# Self Consistency

Self Consistency improves reasoning by generating multiple independent solutions using Chain of Thought and selecting the most frequent or consistent answer.

Instead of trusting a single reasoning path, the model compares several reasoning attempts before deciding on the final answer.

This often improves accuracy on reasoning-heavy tasks.

# Initializing the Language Model

The Groq API is used to access the Large Language Model (LLM). The model selected for this lab is **Llama 3.3 70B Versatile**, which provides strong reasoning capabilities suitable for prompt engineering experiments.

A temperature of **0** is used to make the responses deterministic and consistent, allowing a fair comparison between normal prompting and Chain of Thought prompting.

```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

print("LLM Initialized Successfully!")
```

# Normal Prompting

Normal prompting asks the model to directly answer the question without explicitly requesting intermediate reasoning steps.

This serves as the baseline for comparison with Chain of Thought prompting.

```python
from langchain_core.messages import HumanMessage

question = "If a train travels 60 km in one hour, how far will it travel in 3 hours?"

response = llm.invoke(
    [
        HumanMessage(content=question)
    ]
)

print(response.content)
```

# Chain of Thought Prompting

Chain of Thought (CoT) prompting encourages the model to reason through the problem step by step before producing the final answer.

This technique often improves performance on mathematical and logical reasoning tasks.

```python
cot_prompt = """
Solve the following problem.

Think step by step before giving the final answer.

Question:

If a train travels 60 km in one hour, how far will it travel in 3 hours?
"""

response = llm.invoke([
    HumanMessage(content=cot_prompt)
])

print(response.content)
```

# Comparison

The responses generated using normal prompting and Chain of Thought prompting are compared to observe differences in reasoning quality and answer completeness.

```python
normal_prompt = "If a train travels 60 km in one hour, how far will it travel in 3 hours?"

cot_prompt = """
Solve the following problem.

Think step by step before answering.

If a train travels 60 km in one hour, how far will it travel in 3 hours?
"""

normal_response = llm.invoke(
    [
        HumanMessage(content=normal_prompt)
    ]
)

cot_response = llm.invoke(
    [
        HumanMessage(content=cot_prompt)
    ]
)

print("="*70)
print("NORMAL PROMPT\n")
print(normal_response.content)

print("\n"+"="*70)

print("CHAIN OF THOUGHT PROMPT\n")
print(cot_response.content)
```

# Testing on Multiple Problems

The following experiment evaluates the performance of normal prompting and Chain of Thought prompting on multiple mathematical and logical reasoning questions.

```python
questions = [
    "What is 25 × 18?",
    "A shopkeeper buys an item for $80 and sells it for $100. What is the profit percentage?",
    "If five workers complete a task in ten days, how many days will ten workers take?",
    "A rectangle has length 12 and width 8. What is its area?",
    "If a train travels 90 km in 2 hours, what is its average speed?",
    "Tom is older than Ali. Ali is older than Sara. Who is the oldest?",
    "Find the next number: 2, 4, 8, 16, ?",
    "If today is Monday, what day will it be after 10 days?",
    "A class has 30 students. If 40% are girls, how many boys are there?",
    "What is the sum of the first ten natural numbers?"
]
```

```python
for i, question in enumerate(questions, start=1):

    print("="*80)
    print(f"Question {i}")
    print(question)

    normal = llm.invoke(
        [
            HumanMessage(content=question)
        ]
    )

    cot = llm.invoke(
        [
            HumanMessage(
                content=f"Think step by step before answering.\n\nQuestion:\n{question}"
            )
        ]
    )

    print("\nNormal Prompt:")
    print(normal.content)

    print("\nChain of Thought Prompt:")
    print(cot.content)

    print("\n")
```

# Observations

During the comparison, the following observations were made:

- Chain of Thought prompting provides detailed reasoning before producing the final answer.
- Normal prompting generally produces shorter responses.
- CoT performs better on mathematical and logical reasoning tasks.
- The reasoning process makes the model's decision-making more transparent.
- CoT responses are easier to verify because intermediate steps are shown.
