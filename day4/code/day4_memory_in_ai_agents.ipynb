# Introduction

Large Language Models process each prompt independently unless previous conversation is provided.

Memory enables AI systems to retain context from earlier interactions, allowing conversations to become more natural and coherent.

This notebook demonstrates how conversational memory can be implemented using LangChain concepts and Python.

# Why Memory?

Without memory:

- The AI forgets previous messages.
- Every prompt is treated independently.
- Conversations become repetitive.

With memory:

- Previous context is preserved.
- Responses become more personalized.
- Multi-turn conversations become possible.
- AI assistants behave more naturally.

# Types of Memory

## Short-Term Memory

- Stores information during the current conversation.
- Reset when the session ends.

## Long-Term Memory

- Stores information across multiple sessions.
- Useful for personalized assistants and long-running applications.

````python
from dotenv import load_dotenv
import os

from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, AIMessage

load_dotenv()
````

````python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

print("Groq LLM Initialized Successfully!")
````

# Conversation Without Memory

The following example demonstrates how the model behaves when previous messages are not provided.

````python
response = llm.invoke("My name is Faizan.")

print(response.content)
````

````python
response = llm.invoke("What is my name?")

print(response.content)
````

# Introducing Conversation Memory

One simple way to simulate memory is by storing previous messages in a list.

Each new prompt includes the entire conversation history, allowing the model to maintain context.

````python
conversation = []
````

````python
def chat(user_input):
    conversation.append(HumanMessage(content=user_input))

    response = llm.invoke(conversation)

    conversation.append(AIMessage(content=response.content))

    return response.content
````

# Testing Conversation Memory

The following examples demonstrate how the chatbot remembers previous messages.

````python
print(chat("My name is Faizan."))
````

````python
print(chat("I study Artificial Intelligence at FAST University."))
````

````python
print(chat("What is my name and where do I study?"))
````

# Clearing Conversation Memory

Sometimes it is necessary to start a completely new conversation.

By clearing the conversation history, the chatbot forgets all previous interactions and begins with a fresh context.

````python
conversation.clear()

print("Conversation memory cleared!")
````

# Starting a New Conversation

After clearing the memory, the chatbot should no longer remember any previous information.

````python
print(chat("What is my name?"))
````

Memory enhances the capabilities of conversational AI by maintaining context throughout a conversation.

Advantages include:

- Personalized responses
- Reduced repetition

# Limitations

Although conversational memory is useful, it also has some limitations:

- Longer conversations require more tokens.
- Very large histories can increase response time.
- Long-term memory requires persistent storage.
- Incorrect information stored in memory may affect future responses.
- Memory management strategies are required for production systems.
